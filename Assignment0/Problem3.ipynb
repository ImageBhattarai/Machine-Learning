```python
import numpy as np
```

```python
#Putting values in array
food = np.array([[3,3],[4,2]])
cost = np.array([[11.25],[10]])
```

```python
#Finding inverse of array and them multiplying with cost
np_arr = np.linalg.inv(food)
price = np_arr.dot(cost)
```

```python
print("The cost of soft tacos is " + '{}'.format(price[0][0]))
print("The cost of burritos is " + '{}'.format(price[1][0]))
```
