```python
from sklearn import tree
import pandas as pd
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
#import test and train csv from excel file
data= pd.read_csv('/content/drive/MyDrive/train.csv')  
x_train = data.iloc[:, 1:9].values  
y_train = data.iloc[:, 9].values  

dataa= pd.read_csv('/content/drive/MyDrive/test.csv')  
x_test = dataa.iloc[:, 1:9].values  
y_test = dataa.iloc[:, 9].values  
```

```python
# model
classif = tree.DecisionTreeClassifier(max_depth=2, criterion='entropy')  
classif.fit(x_train, y_train)
```

```python
# decision tree
plt.figure(figsize=(12,8))
tree.plot_tree(classif,  
               feature_names=['Local Price,', 'Bathrooms', 'Land Area,', 'Living Area', 'Num Garages,', 'Num Rooms', 'Num Bedrooms,', 'Age of Home'],
               class_names=['apartment', 'condo', 'house'],
               filled='true')

y_prediction_training = classif.predict(x_train)
y_prediction_test = classif.predict(x_test)

print(f'Training Accuracy- {accuracy_score(y_train, y_prediction_training)}')
print(f'Test Accuracy- {accuracy_score(y_test, y_prediction_test)}')
```
