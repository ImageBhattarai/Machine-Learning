```python
img_arr = np.array(img)
```

```python
print(img_arr.shape)
print(type(img_arr))
```

```python
point1 = [255, 0, 0]
value1 = np.linalg.norm(img_arr - point1)
print(value1)
```

```python
point2 = [0,255,0]
value2 = np.linalg.norm(img_arr - point2)
print(value2)
```

```python
point3 = [0,0,255]
value3 = np.linalg.norm(img_arr - point3)
print(value3)
```

```python
# Point 3 i.e [0,0,255] has the least distance.
img_arr[:] = [0,0,255]
```

```python
cv2.imwrite("output1.png",img_arr)
```
