```python
from curses.ascii import isalpha
import numpy as np
from matplotlib import pyplot as plt
```

```python
file1 = open("./drive/MyDrive/test.txt","r+")
```

```python
my_dict = {}
```

```python
for sentence in file1:
  for alphabet in sentence:
    if alphabet.isalpha():
      if alphabet.lower() in my_dict:
          my_dict[alphabet.lower()] += 1
      else:
          my_dict[alphabet.lower()] = 1
```

```python
print("Unsorted histogram")
print(my_dict)
```

```python
print("\nSorted Histogram")
for key in sorted(my_dict):
    print("%s: %s" % (key, my_dict[key]))

file1.close()
```

```python
import numpy as np
```

```python
file1 = open("./drive/MyDrive/test.txt","r+")
```

```python
a = []
```

```python
for sentence in file1:
  for alphabet in sentence:
    if alphabet.isalpha():
      a.append(alphabet.lower())
```

```python
unique, counts = np.unique(a, return_counts=True)
print("\nHistogram array using numpy")
print(np.asarray((unique,counts)).T)
```

```python
file1.close()
```

```python
# For histogram graphs
```

```python
from matplotlib import pyplot as plt
```

```python
data = np.asarray(counts)
bins = [10,20,30,40,50,60,70]
```

```python
plt.subplot(2,1,1)
plt.hist(data, bins=bins, edgecolor = 'black', density=True )
plt.title('Frequencies of alphabets normazlied')
plt.xlabel('Number of times alphabet repeated')
plt.ylabel('Number of occurences')
```

```python
plt.subplot(2,1,2)
plt.hist(data, bins=bins, edgecolor = 'black')


plt.title('Frequencies of alphabets unnormazlied')
plt.xlabel('Number of times alphabet repeated')
plt.ylabel('Number of occurences')
plt.tight_layout()
plt.show()
```
